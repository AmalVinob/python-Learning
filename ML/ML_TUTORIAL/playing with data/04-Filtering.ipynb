# Filtering: Using Conditionals to Filter Rows and Columns 

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"]
}
```

```python
people = {"name": ["amal","deepika","ship"],
           "age":[23,20,1],
            "mark":[50,60,100]}
df_p = pd.DataFrame(people)
df_p
```

```python
df = pd.DataFrame(people)
```

```python
df_p['name'] == "deepika"
```

```python
df
```

```python
df['last'] == 'Doe'
```

The above return is a series, not a dataframe. You can think of this as a filter mask.

Try assigning return to a variable (note 'filter' is a reserved keyword in python).

```python
filt = (df['last'] == 'Doe')
```

```python
df[filt]
```

Now we get a dataframe! Alternative one line syntax (that _can_ be difficult to read):

```python
df[df['last'] == 'Doe']
```

Another way is to use the .loc indicator on the variable:

```python
df.loc[filt]
```

Using .loc also give you the option of selecting specific columns (which will return a series):

```python
df.loc[filt, 'email']
```

Can't use python built-in 'AND' and 'OR', so need to use symbols '&' and '|'.

```python
filt = (df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
filt = (df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
df.loc[filt, "email"]
```

```python
df.loc[filt, 'email']
```

```python
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
```

```python
df.loc[filt, 'email']
```

Also get the opposite of a filter. For example where we want to get all the rows where last is not Schafer and first isn't John, using the tilde ('~') symbol.

```python
df
```

```python
df.loc[~filt, 'email']
```

Limit to only certain countries. First create a list:

```python
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
```

Now define only those rows whose Country column matches one of those in the list:

```python
filt = df['Country'].isin(countries)
```

Now show the data returned:

```python
df.loc[filt, 'Country']
```
