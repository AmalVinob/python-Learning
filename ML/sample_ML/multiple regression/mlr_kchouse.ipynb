```python
import statsmodels.api as sm

def backwardElimination(x, SL):
    numVars = len(x[0])  # Number of variables/features in the input 'x'
    temp = np.zeros((21613, 19)).astype(int)  # Temporary array to store eliminated features

    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, x).fit()  # Fit OLS model with current set of features 'x'
        maxVar = max(regressor_OLS.pvalues).astype(float)  # Find maximum p-value (least significant)
        adjR_before = regressor_OLS.rsquared_adj.astype(float)  # Get adjusted R-squared before elimination

        if maxVar > SL:  # Check if maximum p-value is greater than significance level 'SL'
            for j in range(0, numVars - i):  # Iterate through each feature
                if regressor_OLS.pvalues[j].astype(float) == maxVar:  # Identify the feature with the max p-value
                    temp[:, j] = x[:, j]  # Store the feature to 'temp'
                    x = np.delete(x, j, 1)  # Remove the identified feature from 'x'

                    tmp_regressor = sm.OLS(y, x).fit()  # Fit OLS model with updated feature set 'x'
                    adjR_after = tmp_regressor.rsquared_adj.astype(float)  # Get adjusted R-squared after elimination

                    if adjR_before >= adjR_after:  # Check if model performance improves
                        x_rollback = np.hstack((x, temp[:, [0, j]]))  # Concatenate 'x' with stored feature
                        x_rollback = np.delete(x_rollback, j, 1)  # Roll back by removing the identified feature
                        print(regressor_OLS.summary())  # Display regression summary
                        return x_rollback  # Return the modified 'x_rollback'
                    else:
                        continue  # Continue to next iteration if model performance doesn't improve
    regressor_OLS.summary()  # Display regression summary (outside loop)
    return x  # Return the final set of selected features
```
