```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([[1, 2, 3], [5, 6, 7]])

fig, ax = plt.subplots()

# Iterate over each row of 'a' to plot each vector
for i in range(len(a)):
    ax.quiver(0, 0, a[i, 0], a[i, 1], angles='xy', scale_units='xy', scale=1, color='r')

ax.set_xlim([-2, 8])  # Adjust the x-axis limits based on the vectors in 'a'
ax.set_ylim([-2, 8])  # Adjust the y-axis limits based on the vectors in 'a'

plt.grid()
plt.show()
```

```python
ax = plt.axes()

ax.arrow(1,2,6,6, head_width = .5, head_length = 1)

plt.ylim(0,10)
plt.xlim(0,10)
plt.show()
```
