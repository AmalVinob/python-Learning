```python
import pandas as pd
from factor_analyzer import FactorAnalyzer

# Assuming 'data' is your DataFrame containing the dataset

# Convert categorical variables to dummy variables using pd.get_dummies()
data = pd.get_dummies(x_filtered)

# Perform factor analysis on numerical variables
# Assuming 'numerical_data' contains only numerical columns
numerical_data = data.select_dtypes(include=['int64'])

# Initialize FactorAnalyzer with desired parameters
n_factors = 5  # Choose the number of factors
fa = FactorAnalyzer(n_factors, rotation=None)  # You can choose a rotation method (e.g., 'varimax')

# Fit the model
fa.fit(numerical_data)

# Get factor loadings
factor_loadings = fa.loadings_

# Get eigenvalues and variance explained
eigenvalues, variance_explained = fa.get_eigenvalues()

# Print factor loadings and eigenvalues
print("Factor Loadings:\n", factor_loadings)
print("\nEigenvalues:\n", eigenvalues)
print("\nVariance Explained:\n", variance_explained)
```

```python
import matplotlib.pyplot as plt

# Assuming 'eigenvalues' contains the eigenvalues obtained from factor analysis
# 'eigenvalues' is a list or array of eigenvalues in descending order

# Plotting the scree plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='-')
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalues')
plt.axhline(y=1.0, color='r', linestyle='--')  # Optional: Draw a horizontal line at eigenvalue = 1
plt.grid(True)
plt.show()
```
